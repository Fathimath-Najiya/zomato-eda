```python
print(data["highlights"].unique())
```

```python
hl = []
data["highlights"].apply(lambda x : hl.extend(x[2:-2].split("', '")))

hl = pd.Series(hl)
print("Total number of unique highlights = ", hl.nunique())
```

* **hl = [ ]**: This initializes an empty list called hl to store individual highlights.


* **x[2:-2]** slices the string x to remove the first two and last two characters, effectively removing the square brackets.
.**split("', '")** splits the resulting string into a list of highlights using **', '** as the separator. This assumes that the highlights are enclosed in single quotes and separated by **', '**.
* The **hl.extend()** method is used to add the individual highlights to the hl list.

# **EXPLORATORY DATA ANALYSIS(EDA)**

* there may be many outlets for a restaurant
* So first divide it into  chain and single
  * chain----> more than 1 outlet
  * single---> only 1 outlet

```python
outlet=data["name"].value_counts()
print(outlet)
```

```python
chains = outlet[outlet >= 2]
single = outlet[outlet == 1]
```

```python
chains
```

```python
print("Total Restaurants = ", data.shape[0])
print("Total Restaurants that are part of some chain = ", data.shape[0] - single.shape[0])#total-singles
print("Total Restaurants that are not part of some chain = ",data.shape[0]-chains.shape[0])
```

```python
chains.head(10)
```

```python
top10_chains = data["name"].value_counts()[:10].sort_values(ascending=True)
print(top10_chains)
```

```python
height = top10_chains.values
bars = top10_chains.index
y_pos = np.arange(len(bars))

fig = plt.figure(figsize=[11,7], frameon=False)
ax = fig.gca()
ax.spines["top"].set_visible("#424242")
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#424242")
ax.spines["bottom"].set_color("#424242")

colors = ["green","blue","magenta","cyan","gray","yellow","purple","violet","orange","red","maroon"]
random.shuffle(colors)
#colors = ["#f9cdac","#f2a49f","#ec7c92","#e65586","#bc438b","#933291","#692398","#551c7b","#41155e","#2d0f41"]
plt.barh(y_pos, height, color=colors)

plt.xticks(color="#424242")

plt.yticks(y_pos, bars, color="#424242")
plt.xlabel("Number of outlets in India")

for i, v in enumerate(height):
    ax.text(v+3, i, str(v), color='#424242')
plt.title("Top 10 Restaurant chain in India (by number of outlets)")


plt.show()
```

**From this chart It is clear that majorly dominaed by big fast food chains**

```python
atleast_5_outlets = outlet[outlet > 5]
top10_chains2 = data[data["name"].isin(atleast_5_outlets.index)].groupby("name").mean()["aggregate_rating"].sort_values(ascending=False)[:10].sort_values(ascending=True)
height = pd.Series(top10_chains2.values).map(lambda x : np.round(x, 2))#round the values
bars = top10_chains2.index
y_pos = np.arange(len(bars))

fig = plt.figure(figsize=[11,7], frameon=False)
ax = fig.gca()
ax.spines["top"].set_visible("#424242")
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#424242")
ax.spines["bottom"].set_color("#424242")

#colors = ["green","blue","magenta","cyan","gray","yellow","purple","violet","orange","red","maroon"]
#random.shuffle(colors)
colors = ["#f9cdac","#f2a49f","#ec7c92","#e65586","#bc438b","#933291","#692398","#551c7b","#41155e","#2d0f41"]
#colors = ['#fded86', '#fce36b', '#f7c65d', '#f1a84f', '#ec8c41', '#e76f34', '#e25328', '#b04829', '#7e3e2b', '#4c3430']
plt.barh(y_pos, height, color=colors)

plt.xlim(3)
plt.xticks(color="#424242")
plt.yticks(y_pos, bars, color="#424242")
plt.xlabel("Number of outlets in India")

for i, v in enumerate(height):
    ax.text(v + 0.01, i, str(v), color='#424242')#showing ratings
plt.title("Top 10 Restaurant chain in India (by average Rating)")


plt.show()
```

here no fast food chain appears in this chart. To maintain a high rating,
restaurants need to provide the best services

# **RATING AND COST**

```python
sns.kdeplot(data['aggregate_rating'], shade=True)
plt.title("Ratings distribution")
plt.show()
```

There is a huge spike at **0** which might account for newly opened or unrated restaurants.Majority restaurant has rating in between **3-4**

```python
sns.kdeplot(data['average_cost_for_two'], shade=True)
plt.xlim([0, 6000])
plt.xticks(range(0,6000,500))
plt.title("Average cost for 2 distribution")
plt.show()
```

majority of restaurants are budget friendly

```python
pr_count = data.groupby("price_range").count()["name"]
#print(pr_count)

fig = plt.figure(figsize=[10,6], frameon=False)
ax = fig.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#424242")
ax.spines["bottom"].set_color("#424242")

#colors = ["green","blue","magenta","cyan","gray","yellow","purple","violet","orange","red","maroon"]
#random.shuffle(colors)
colors = ['#11174b', '#2a5599', '#3e94c0', '#78c6d0', '#dcecc9']
plt.bar(pr_count.index, pr_count.values, color=colors)

plt.xticks(range(0, 5), color="#424242")
plt.yticks(range(0, 40000, 5000), color="#424242")
plt.xlabel("Price Ranges")

for i, v in enumerate(pr_count):
    ax.text(i+0.9, v+500, str(v), color="#424242")
plt.title("Number of restaurant(by Price range)")


plt.show()
```

Price range chart supports our previous observation from the Average cost chart. Number of restaurant decreases with increase in price range.



```python
sns.barplot(x='price_range', y='aggregate_rating', data=data)
plt.ylim(1)
plt.title("Relationship between Price range and Ratings")
plt.show()
```

* The rating of higher price range is above **3.5**
* From this it i clear that higher price range restaurant provides good services hence more chances for getting good rating
